```python
# ===== Internship Practice - Simple Version =====


# Task 1: Login Check
print("Task 1: Login Check")

correct_user = "admin"
correct_pass = "1234"

user = "admin"
password = "1234"

if user == correct_user and password == correct_pass:
    print("Login Success")
else:
    print("Login Failed")

print()


# Task 2: Pass/Fail Count
print("Task 2: Result Count")

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Passed:", pass_count)
print("Failed:", fail_count)

print()


# Task 3: Clean Names
print("Task 3: Clean Names")

names = [" Alice ", "bob", " CHARLIE "]
clean_names = []

for n in names:
    clean_names.append(n.strip().lower())

print(clean_names)

print()


# Task 4: Message Length
print("Task 4: Message Length")

texts = ["Hi", "Welcome to the platform", "OK"]

for t in texts:
    if len(t) > 10:
        print(t, "- Long")
    else:
        print(t, "- Short")

print()


# Task 5: Count Errors
print("Task 5: Error Count")

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

count = 0
for log in logs:
    if log == "ERROR":
        count += 1

print("Errors:", count)


```
