```python
#problem-1
sentence = "python is easy and python is powerful"

words = sentence.split()
unique = set(words)

print("Unique words count:", len(unique))
print("Unique words:", unique)





#problem-2
employees = {"Ravi": 75000, "Anita": 68000, "Kiran": 72000}

name = max(employees, key=employees.get)
print("Highest Salary:", name, "-", employees[name])





#problem-3
numbers = [45, 22, 89, 10, 66]

print("List:", numbers)
print("Max:", max(numbers))
print("Min:", min(numbers))





#problem-4
prices = [450, 1200, 899, 1500, 300]

count = sum(1 for p in prices if p > 1000)
print("Products above 1000:", count)





#problem-5
attendance = ["P", "P", "A", "P", "P"]

percentage = (attendance.count("P") / len(attendance)) * 100
print("Attendance Percentage:", percentage)





#problem-6

phone_numbers = [9876543210, 9123456789, 9876543210]

print("Unique phone numbers:", set(phone_numbers))





#problem-7
text = "python"

freq = {}
for ch in text:
    freq[ch] = freq.get(ch, 0) + 1

print(freq)





#problem-8
data = [10, 20, 30]

print("Tuple:", tuple(data))





#problem-9
employee = {"name": "Ravi", "salary": 50000}

if "name" in employee:
    print("Employee exists")





#problem-10


marks = [70, 75, 80, 65]

average = sum(marks) / len(marks)
print("Average Marks:", average)
```
