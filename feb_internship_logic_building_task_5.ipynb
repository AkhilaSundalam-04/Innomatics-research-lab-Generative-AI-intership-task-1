```python
#problem-1:Social Media – Post Engagement Analyzer
def analyze_post_engagement(likes_list):
    total_likes = 0
    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)


analyze_post_engagement([200, 300, 250, 300])

#Problem-2: Healthcare – Medicine Stock Alert System
def check_medicine_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)


check_medicine_stock(6)

#Problem 3: Agriculture – Rainfall Adequacy Checker
def check_rainfall(rainfall_list, required_level):
    total = 0
    for rain in rainfall_list:
        total += rain

    average = total / len(rainfall_list)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)


check_rainfall([60, 70, 80, 75], 70)

#Problem 4: Social Media – Duplicate Account Detection
def detect_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)


detect_duplicates(["akhila", "ram", "akhila", "john"])

#Problem 5: Healthcare – Appointment Eligibility Checker
def check_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)


check_eligibility(21)

#Problem 6: Agriculture – Premium Crop Price Filter
def filter_premium_crops(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)


filter_premium_crops([1500, 2500, 3200, 1800])

#Problem 7: System Monitoring – Application Health Checker
def check_system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)


check_system_health(7)

#Problem 8: Banking – Daily Transaction Limit Checker
def check_transaction(amount):
    limit = 50000

    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


check_transaction(60000)

#Problem 9: E-Learning – Student Attendance Eligibility System
def check_attendance(attendance_list):
    total_classes = len(attendance_list)
    attended = 0

    for status in attendance_list:
        if status == 1:
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)


check_attendance([1, 1, 0, 1, 1])

#Problem 10: Smart Electricity Bill Analyzer
def calculate_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Bill:", bill)
    print("Usage Level:", usage)


calculate_bill(250)
```
